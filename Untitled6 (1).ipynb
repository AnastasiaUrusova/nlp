```python
import nltk
import re

nltk.download("stopwords") # поддерживает удаление стоп-слов
nltk.download('punkt') # делит текст на список предложений
nltk.download('wordnet') # проводит лемматизацию
from nltk.corpus import stopwords
```

```python
html_pattern = re.compile('<.*?>')
df['review'] = df['review'].replace(html_pattern, '')
df
```

```python
df["review_t"]=df["review"].str.replace("[</]","",regex=True)
df
```

```python
import string
string.punctuation
```

```python
def remove_punct(text):
    text_nopunct="".join([char for char in text if char not in string.punctuation])
    return text_nopunct
df["review_clean"]=df["review_t"].apply(lambda x: remove_punct(x))
df.head()
```

```python
import re
def tokenize(text):
    tokenize=re.split("\W+",text)
    return tokenize
df["review_tokenized"]=df["review_clean"].apply(lambda x: tokenize(x.lower()))
df.head()
```

```python
import nltk
stopword=nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
  text=[word for word in tokenized_list if word not in stopword]
  return text
df["review_nostop"]=df["review_tokenized"].apply(lambda x: remove_stopwords(x))
df.head()
```
